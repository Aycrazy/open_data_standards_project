```python
def ckan_to_df(soup):
    new_dict ={}
    for d in json.loads(str(soup))['result']['records']:
        for key in  json.loads(str(soup))['result']['records'][0].keys():
            if key in new_dict:
                new_dict[key].append(d[key])
            else: 
                new_dict[key]=[d[key]]
    return pd.DataFrame(new_dict)
```
